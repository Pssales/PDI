# Transformação de imagens

Verificando versão instalada do OpenCV

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
```

## Suavização de imagens

Filtros passa-baixa

```python
imagem = cv2.imread("input/ruido2.png")
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

plt.imshow(imagem)
plt.title("Original");
```

```python
kernel = np.ones((3,3), np.float32)/9
kernel
```

```python
img_filtro1 = cv2.filter2D(imagem, -1, kernel)

plt.imshow(img_filtro1)
plt.title("Suavização 3x3 - 1º filtro");
```

```python
kernel = np.array([[1, 2,1], [2, 4, 2], [1,2,1]])*1/16
kernel
```

```python
img_filtro2 = cv2.filter2D(imagem, -1, kernel)

plt.imshow(img_filtro2)
plt.title("Suavização 3x3 - 2º filtro");
```

```python
kernel = np.array([[1, 2 ,1], [1, 2, 1], [1,1,1]])*1/10
kernel
```

```python
img_filtro3 = cv2.filter2D(imagem, -1, kernel)

plt.imshow(img_filtro2)
plt.title("Suavização 3x3 - 3º filtro");
```

```python
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.imshow(imagem)
plt.title("Original")

plt.subplot(2,2,2)
plt.imshow(img_filtro1)
plt.title("Suavização Filtro 1")

plt.subplot(2,2,3)
plt.imshow(img_filtro2)
plt.title("Suavização Filtro 2")

plt.subplot(2,2,4)
plt.imshow(img_filtro3)
plt.title("Suavização Filtro 3");
```

Diferença entre a imagem original e os filtros

```python
dif_f1 = imagem - img_filtro1
dif_f2 = imagem - img_filtro2
dif_f3 = imagem - img_filtro3
```

```python
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.imshow(imagem)
plt.title("Original")

plt.subplot(2,2,2)
plt.imshow(dif_f1)
plt.title("Difrença original - filtro 1")

plt.subplot(2,2,3)
plt.imshow(dif_f2)
plt.title("Difrença original - Filtro 2")

plt.subplot(2,2,4)
plt.imshow(dif_f3)
plt.title("Difrença original - Filtro 3");
```

### Filtro Média e Mediana

```python
imagem_suav_padrao = cv2.blur(imagem, (5,5))

plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.imshow(imagem_suav_padrao)
plt.title("Suavização")


imagem_suav_gaussian = cv2.GaussianBlur(imagem, (7,7), 0)

plt.subplot(2,2,2)
plt.imshow(imagem_suav_gaussian)
plt.title("Suavização Gaussiana")

imagem_suav_mediana = cv2.medianBlur(imagem, 5)
            
plt.subplot(2,2,3)
plt.imshow(imagem_suav_mediana)
plt.title("Suavização Mediana")

imagem_suav_bilateral = cv2.bilateralFilter(imagem, 9, 75, 75)

plt.subplot(2,2,4)
plt.imshow(imagem_suav_bilateral)
plt.title("Suavização Bilateral")
```
