# Transformação de imagens

Importação das bibliotecas que serão utilizadas 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2

from skimage.io import imread
from skimage.filters import roberts, sobel, scharr, prewitt
```

## Suavização de imagens

Filtros passa-baixa

```python
imagem = cv2.imread("input/ruido2.png")
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

plt.imshow(imagem)
plt.title("Original");
```

```python
kernel = np.ones((3,3), np.float32)/9
kernel
```

```python
img_filtro1 = cv2.filter2D(imagem, -1, kernel)

plt.imshow(img_filtro1)
plt.title("Suavização 3x3 - 1º filtro");
```

```python
kernel = np.array([[1, 2,1], [2, 4, 2], [1,2,1]])*1/16
kernel
```

```python
img_filtro2 = cv2.filter2D(imagem, -1, kernel)

plt.imshow(img_filtro2)
plt.title("Suavização 3x3 - 2º filtro");
```

```python
kernel = np.array([[1, 2 ,1], [1, 2, 1], [1,1,1]])*1/10
kernel
```

```python
img_filtro3 = cv2.filter2D(imagem, -1, kernel)

plt.imshow(img_filtro2)
plt.title("Suavização 3x3 - 3º filtro");
```

```python
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.imshow(imagem)
plt.title("Original")

plt.subplot(2,2,2)
plt.imshow(img_filtro1)
plt.title("Suavização Filtro 1")

plt.subplot(2,2,3)
plt.imshow(img_filtro2)
plt.title("Suavização Filtro 2")

plt.subplot(2,2,4)
plt.imshow(img_filtro3)
plt.title("Suavização Filtro 3");
```

Diferença entre a imagem original e os filtros

```python
dif_f1 = imagem - img_filtro1
dif_f2 = imagem - img_filtro2
dif_f3 = imagem - img_filtro3
```

```python
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.imshow(imagem)
plt.title("Original")

plt.subplot(2,2,2)
plt.imshow(dif_f1)
plt.title("Difrença original - filtro 1")

plt.subplot(2,2,3)
plt.imshow(dif_f2)
plt.title("Difrença original - Filtro 2")

plt.subplot(2,2,4)
plt.imshow(dif_f3)
plt.title("Difrença original - Filtro 3");
```

### Filtro Média e Mediana

```python
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.imshow(imagem)
plt.title("Original")

imagem_mediana = cv2.medianBlur(imagem, 5)
            
plt.subplot(2,2,2)
plt.imshow(imagem_mediana)
plt.title("Suavização Mediana");
```

```python
dif_mediana = imagem - imagem_mediana
plt.imshow(dif_mediana)
plt.title("Difrença original - mediana");
```

## Bordas

```python
img = imread("input/ruido.jpg", as_gray=True)
op_roberts = roberts(img)
op_sobel = sobel(img)

diff_ori_roberts = img - op_roberts
diff_ori_sobel = img - op_sobel
```

```python
plt.imshow(img, cmap=plt.cm.gray)
plt.title("Original");
```

```python
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.imshow(op_roberts, cmap=plt.cm.gray)
plt.title("Operador de Roberts")

plt.subplot(2,2,2)
plt.imshow(diff_ori_roberts, cmap=plt.cm.gray)
plt.title("Diferença Original - Roberts")

plt.subplot(2,2,3)
plt.imshow(op_sobel, cmap=plt.cm.gray)
plt.title("Operador de Sobel");

plt.subplot(2,2,4)
plt.imshow(diff_ori_sobel, cmap=plt.cm.gray)
plt.title("Diferença Original - Sobel");
```
