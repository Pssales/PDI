```python
import numpy as np
import matplotlib.pyplot as plt

from skimage.io import imread
from skimage.filters import roberts, sobel, scharr, prewitt

import cv2
```

```python
img = imread("input/ruido.jpg", as_gray=True)
op_roberts = roberts(img)
op_sobel = sobel(img)
```

```python
diff_ori_roberts = img - op_roberts
diff_ori_sobel = img - op_sobel
```

```python
max_diff = np.max(np.maximum(diff_ori_roberts, diff_ori_sobel))
max_diff
```

```python
fig, axes = plt.subplots(nrows=4, ncols=2, sharex=True, sharey=True, figsize=(12, 12))

ax = axes.ravel()

ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title('Imagem Original')

ax[2].imshow(op_roberts, cmap=plt.cm.gray)
ax[2].set_title('Operador de Roberts')

ax[3].imshow(diff_ori_roberts, cmap=plt.cm.gray, vmax=max_diff)
ax[3].set_title('Diferença Original - Roberts')

ax[4].imshow(op_sobel, cmap=plt.cm.gray)
ax[4].set_title('Operador de Sobel')

ax[5].imshow(diff_ori_sobel, cmap=plt.cm.gray, vmax=max_diff)
ax[5].set_title('Diferença Original - Sobel')


for a in ax:
    a.axis('off')

plt.tight_layout()
plt.show()
```

## Suavização de imagens

A suavização de imagens (ou Blurring) auxilia no processo de remoção de ruídos e diminuição de detalhes de uma imagem.

```python
imagem = cv2.imread("input/ruido.jpg")
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

plt.imshow(imagem)
plt.title("Idoso")
```

```python
kernel = np.ones((3,3), np.float32)/9
imagem_suavizada = cv2.filter2D(imagem, -1, kernel)

plt.imshow(imagem_suavizada)
plt.title("Suavização 3x3")
```

```python
kernel = np.ones((7,7), np.float32)/49
imagem_suavizada = cv2.filter2D(imagem, -1, kernel)

plt.imshow(imagem_suavizada)
plt.title("Suavização 7x7")
```

Outras operações de suavização com filtros especializados.

```python
imagem = cv2.imread("input/ruido.jpg")
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

plt.imshow(imagem)
plt.title("Suavização 7x7")
```

```python
imagem = cv2.imread("input/ruido.jpg")
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

plt.imshow(imagem)
plt.title("Original")
```

```python
imagem_suav_padrao = cv2.blur(imagem, (5,5))

plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.imshow(imagem_suav_padrao)
plt.title("Suavização")


imagem_suav_gaussian = cv2.GaussianBlur(imagem, (7,7), 0)

plt.subplot(2,2,2)
plt.imshow(imagem_suav_gaussian)
plt.title("Suavização Gaussiana")

imagem_suav_mediana = cv2.medianBlur(imagem, 5)
            
plt.subplot(2,2,3)
plt.imshow(imagem_suav_mediana)
plt.title("Suavização Mediana")

imagem_suav_bilateral = cv2.bilateralFilter(imagem, 9, 75, 75)

plt.subplot(2,2,4)
plt.imshow(imagem_suav_bilateral)
plt.title("Suavização Bilateral")
```
