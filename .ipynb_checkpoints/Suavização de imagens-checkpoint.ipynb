## Suavização de imagens

Filtros passa-baixa

Importação das bibliotecas que serão utilizadas 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2

from skimage.io import imread
from skimage.filters import roberts, sobel, scharr, prewitt
```

Leitura da Imagem

```python
imagem = cv2.imread("input/ruido2.png")
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

plt.imshow(imagem)
plt.title("Original");
```

### Primeiro Filtro

Definindo máscara:
<img src="C:/Users/Camila/Desktop/PDI/mascaras/primeira.jpg" style="height:100px">

```python
kernel = np.ones((3,3), np.float32)/9
kernel
```

```python
img_filtro1 = cv2.filter2D(imagem, -1, kernel)

plt.imshow(img_filtro1)
plt.title("1º filtro");
```

### Segundo Filtro

Definindo máscara:
<img src="C:/Users/Camila/Desktop/PDI/mascaras/segunda.jpg" style="height:100px">

```python
kernel = np.array([[1, 2,1], 
                   [2, 4, 2], 
                   [1,2,1]])*1/16
kernel
```

```python
img_filtro2 = cv2.filter2D(imagem, -1, kernel)

plt.imshow(img_filtro2)
plt.title("2º filtro");
```

### Terceiro Filtro

Definindo máscara:
<img src="C:/Users/Camila/Desktop/PDI/mascaras/terceira.jpg" style="height:100px">

```python
kernel = np.array([[1, 2 ,1], 
                   [1, 2, 1], 
                   [1,1,1]])*1/10
kernel
```

```python
img_filtro3 = cv2.filter2D(imagem, -1, kernel)

plt.imshow(img_filtro2)
plt.title("3º filtro");
```

```python
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.imshow(imagem)
plt.title("Original")

plt.subplot(2,2,2)
plt.imshow(img_filtro1)
plt.title("Filtro 1")

plt.subplot(2,2,3)
plt.imshow(img_filtro2)
plt.title("Filtro 2")

plt.subplot(2,2,4)
plt.imshow(img_filtro3)
plt.title("Filtro 3");
```

Diferença entre a imagem original e os filtros

```python
dif_f1 = imagem - img_filtro1
dif_f2 = imagem - img_filtro2
dif_f3 = imagem - img_filtro3
```

```python
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.imshow(imagem)
plt.title("Original")

plt.subplot(2,2,2)
plt.imshow(dif_f1)
plt.title("Original - filtro 1")

plt.subplot(2,2,3)
plt.imshow(dif_f2)
plt.title("Original - Filtro 2")

plt.subplot(2,2,4)
plt.imshow(dif_f3)
plt.title("Original - Filtro 3");
```

### Filtro Mediana

```python
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.imshow(imagem)
plt.title("Original")

imagem_mediana = cv2.medianBlur(imagem, 5)
            
plt.subplot(2,2,2)
plt.imshow(imagem_mediana)
plt.title("Mediana");
```

```python
dif_med = imagem - imagem_mediana

plt.imshow(dif_med)
plt.title("Original - Mediana");
```
